```python
import pandas as pd
df=pd.read_csv("census.csv")
df.head(10)
```

```python
df.drop(['education_level','education-num','capital-gain','capital-loss','hours-per-week','native-country','marital-status','age'],axis=1,inplace=True)
df
```

```python
inputs=df.drop('income',axis='columns')
target=df['income']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python

le_workclass=LabelEncoder()
le_occupation=LabelEncoder()
le_relationship=LabelEncoder()
le_race=LabelEncoder()
le_sex=LabelEncoder()
```

```python

inputs['workclass_n']=le_workclass.fit_transform(inputs['workclass'])
inputs['occupation_n']=le_workclass.fit_transform(inputs['occupation'])
inputs['relationship_n']=le_workclass.fit_transform(inputs['relationship'])
inputs['race_n']=le_workclass.fit_transform(inputs['race'])
inputs['sex_n']=le_workclass.fit_transform(inputs['sex'])
inputs.head(10)
```

```python
inputs_n=inputs.drop(['workclass','occupation','relationship','race','sex'],axis='columns')
inputs_n.head(10)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,7,1,2,0]])
```
